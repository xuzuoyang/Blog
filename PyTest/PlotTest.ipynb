```python
import numpy as np
import matplotlib as mlb
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(1000)
data = np.random.standard_normal(25)
plt.plot(data.cumsum(), 'b', lw=0.5)
plt.plot(data.cumsum(), 'm.')
plt.grid(True)
plt.legend(['a simple legend'])
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Cumsum Plot')
plt.xlim(-1, 25)
plt.ylim(np.min(data.cumsum()) - 1, 
         np.max(data.cumsum()) + 1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 10, 1000) #linspace(start, end, num of elements)
y1 = np.sin(x)
y2 = np.cos(x**2)
plt.figure(figsize=(8, 4))
plt.plot(x, y1, 'r', lw=1.5, label='y=sin(x)')
plt.plot(x, y2, 'b--', label='y=cos(x^2)')
plt.axis('scaled')
plt.ylim(-1.2, 1.2)
plt.xlabel('time(s)')
plt.ylabel('volt')
plt.title('time-volt')
plt.legend(loc=0)
plt.show()
```

```python
import numpy as np
import matplotlib as mlb
import matplotlib.pyplot as plt
%matplotlib inline

channel = ['appstore', 'huawei', 'yingyongbao',
           'xiaomi', 'w.others']
register = [14, 8, 4, 3, 2]
sh_open_account = [4, 3, 3, 4, 3]
bind_card = [15, 3, 2, 1, 9]
first_invest = [2, 2, 3, 2, 0]
y_pos = np.arange(len(channel))

plt.figure(figsize=(10,4))
plt.suptitle('user stat by channel')

plt.subplot(221)
plt.barh(y_pos, register, color='b', lw=0.5, label='register user')
plt.yticks(y_pos, channel)
#plt.xlabel('channel')
plt.ylabel('user num')
plt.xlim(0, np.max(register)*3/2)

plt.subplot(222)
plt.barh(y_pos, sh_open_account, color='b', lw=0.5, label='sh open account user')
plt.yticks(y_pos, channel)
#plt.xlabel('channel')
#plt.ylabel('user num')
plt.xlim(0, np.max(sh_open_account)*3/2)

plt.subplot(223)
plt.barh(y_pos, bind_card, color='b', lw=0.5, label='bind card user')
plt.yticks(y_pos, channel)
plt.xlabel('channel')
plt.ylabel('user num')
plt.xlim(0, np.max(bind_card)*3/2)

plt.subplot(224)
plt.barh(y_pos, first_invest, color='b', lw=0.5, label='first invest user')
plt.yticks(y_pos, channel)
plt.xlabel('channel')
#plt.ylabel('user num')
plt.xlim(0, np.max(first_invest)*3/2)

plt.show()
```
