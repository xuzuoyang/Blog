```python
import numpy as np
import matplotlib as mlb
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(1000)
data = np.random.standard_normal(25)
plt.plot(data.cumsum(), 'b', lw=0.5)
plt.plot(data.cumsum(), 'm.')
plt.grid(True)
plt.legend(['a simple legend'])
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Cumsum Plot')
plt.xlim(-1, 25)
plt.ylim(np.min(data.cumsum()) - 1, 
         np.max(data.cumsum()) + 1)
plt.show()
```
